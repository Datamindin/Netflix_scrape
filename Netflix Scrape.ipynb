```python
try:
    r = requests.get('https://www.netflix.com/in/browse/genre/34399')
    r.raise_for_status()
    
    soup = BeautifulSoup(r.text, 'html.parser')
    
    head = soup.find('div', class_='nm-collections-page').find_all('section')
#     print(head)
#     print(len(head))


#     print(head[0])
    for i in head:
#         print(i)
        
#         i.find()
        title = i.find('span', class_ = 'nm-collections-row-name')
        
        if title == None:
            title1 = i.find('h2', class_ = 'nm-collections-row-name')
            if title1 == None:
                pass
            else:
                print(title1.text,":")
                movies = i.find('div', class_='nm-content-horizontal-row').find_all('li')
                for movie in movies:
                    list= movie.find('span',class_ = 'nm-collections-title-name').text
                    print(list)
        else:
            print(title.text,":")
            
            movies = i.find('div', class_='nm-content-horizontal-row').find_all('li')
    
            for movie in movies:
                list= movie.find('span',class_ = 'nm-collections-title-name').text
                print(list)
            
#                 movies = i.find('h2', class_='nm-collections-row').find_all('li')
            
        print("\n")
   
        
        
        
    
    
    
except Exception as e:
    print(e)
```
